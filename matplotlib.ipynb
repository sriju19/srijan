```python
#bar plot 
langs=['c','java','c++']
members=[10,20,34]
plt.bar(langs,members)
plt.show()
```

```python
subjects=["maths", "lspd", "english", "physics", "ds"]
width=0.25
indices=np.arange(len(subjects))
swe_marks=[20, 19, 18, 15, 19]
srij_marks=[20, 20, 15, 16, 20]
plt.bar(indices-width, swe_marks, width=width, label="sweth marks", color="red")
plt.bar(indices, srij_marks, width=width, label="srij marks", color="green")
plt.title("mid 2 marks")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()


plt.xticks(ticks=indices, labels=subjects)


plt.tight_layout()
plt.show()
```

```python

#pie plot
slices=[1433783686, 1366417754, 329064917, 370625568, 216565318]
labels=["china", "india", "united states", "indoensia", "pakistan"]
explode=[0, 0, 0.2, 0, 0]
colors=['orange', "pink", "red", "yellow"]
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, wedgeprops={"edgecolor":"black", "linewidth": 1},colors=colors)
plt.title("pie chat on population")
plt.grid(True)
plt.show()
```

```python
#scatter plot
x=[1,3,5,6,7,2]
y=[2,4,6,54,3,50]
x1=[1,3,5,6,8,6]
y1=[2,4,6,7,9,0]
plt.scatter(x,y,label="high income",color='r')
plt.scatter(x1,y1,label="low income",color='g')
plt.title("random")
plt.xlabel("saving*100")
plt.ylabel("saving*1000")
plt.legend()
plt.show()
```
