```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill'])
plt.show()
#for warning reduction
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
```

```python
sns.distplot(tips['total_bill'], hist=False)
#hist=false
plt.show()
```

```python
#for points using jointplot
sns.jointplot(x='total_bill', y='tip', data=tips, color='orange')
plt.show()
```

```python
#scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips, color='orange')
plt.show()
```

```python
#stripplot useed for varies information in days and the total bill is more on which days

sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, jitter=True)
plt.show()
```

```python
#swarm plot
sns.swarmplot(x='day', y='tip', data=tips)
plt.show()
```

```python
#box plot
sns.boxplot(tips["total_bill"], data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```
